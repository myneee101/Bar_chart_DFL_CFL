```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [1, 122],
    [2, 123],
    [4, 123.8],
    [6, 124.5],
    [8, 126],
    [10, 127.4],
    [12, 128],
    [14, 129.6],
    [16, 130],
    [18, 131.4],
    [20, 132.1],
])
data2 = np.array([
    [1, 180],
    [2, 181],
    [4, 182.8],
    [6, 183.5],
    [8, 184],
    [10, 185.4],
    [12, 186],
    [14, 187.6],
    [16, 188],
    [18, 189.4],
    [20, 190.1],
])
data3 = np.array([
    [1, 240],
    [2, 240.5],
    [4, 241.8],
    [6, 242.5],
    [8, 243],
    [10, 244.4],
    [12, 245],
    [14, 246.6],
    [16, 247],
    [18, 248.4],
    [20, 249.1],
])

x, y = data.T
x, y2 = data2.T
x, y3 = data3.T

fig, ax = plt.subplots()

colors = ['#24b1d1', '#ae24d1', '#24d14d']
bottom = np.zeros(len(x))

ax.bar(x, y, color=colors[0], label='20 cells')
bottom += y
ax.bar(x, y2, bottom=bottom, color=colors[1], label='30 cells')
bottom += y2
ax.bar(x, y3, bottom=bottom, color=colors[2], label='40 cells')

totals = np.array([sum(val) for val in zip(y, y2, y3)])
y_offset = 4
for i, total in enumerate(totals):
    ax.text(x[i], total + y_offset, round(total), ha='center', weight='bold')

# Annotations inside the bars
y_offset = -15
for bar, y_val in zip(ax.patches, totals):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + y_offset,
        round(y_val),
        ha='center',
        color='w',
        weight='bold',
        size=8
    )

ax.set_xlabel("Number of Drivers")
ax.set_ylabel("Total Decryption time in ms")
ax.legend()

plt.grid()
plt.savefig("decryp_stacked_bar_with_legend.pdf", dpi=1200)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [1, 122, 180, 240],
    [2, 123, 181, 240.5],
    [4, 123.8, 182.8, 241.8],
    [6, 124.5, 183.5, 242.5],
    [8, 126, 184, 243],
    [10, 127.4, 185.4, 244.4],
    [12, 128, 186, 245],
    [14, 129.6, 187.6, 246.6],
    [16, 130, 188, 247],
    [18, 131.4, 189.4, 248.4],
    [20, 132.1, 190.1, 249.1],
])

x, y1, y2, y3 = data.T

bar_width = 1.5

fig, ax = plt.subplots(figsize=(12, 6))  # Increase the figure size

bar1 = ax.bar(x, y1, bar_width, label='20 cells', edgecolor='black', color='skyblue')
bar2 = ax.bar(x, y2, bar_width, bottom=y1, label='30 cells', edgecolor='black', color='lightcoral')
bar3 = ax.bar(x, y3, bar_width, bottom=y1 + y2, label='40 cells', edgecolor='black', color='lightgreen')

ax.set_xlabel("Number of Drivers")
ax.set_ylabel("Total Decryption Time (ms)")
ax.legend(loc='upper left')

ax.set_xticks(x)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each part of the bar plot with its value
for i in range(len(x)):
    ax.text(x[i], y1[i] / 2, f'{y1[i]:.1f}', ha='center', va='center', fontsize=10)
    ax.text(x[i], y1[i] + y2[i] / 2, f'{y2[i]:.1f}', ha='center', va='center', fontsize=10)
    ax.text(x[i], y1[i] + y2[i] + y3[i] / 2, f'{y3[i]:.1f}', ha='center', va='center', fontsize=10)

ax.set_title("Decryption Time by Number of Drivers", fontsize=14)
ax.set_xlim(0.5, max(x) + 0.5)  # Adjust the x-axis limits for better presentation

# Add a horizontal line to highlight a specific value, e.g., a threshold
threshold = 240
ax.axhline(threshold, color='red', linestyle='--', label=f'Threshold ({threshold} ms)')
ax.legend()

# Add a shaded region to emphasize a specific range of values
ax.fill_between(x, 220, 260, color='gray', alpha=0.3, label='Acceptable Range')

plt.savefig("advanced_stacked_bar_plot.png", dpi=300, bbox_inches='tight')  # Save as a high-resolution image

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [1, 122, 180, 240],
    [2, 123, 181, 240.5],
    [4, 123.8, 182.8, 241.8],
    [6, 124.5, 183.5, 242.5],
    [8, 126, 184, 243],
    [10, 127.4, 185.4, 244.4],
    [12, 128, 186, 245],
    [14, 129.6, 187.6, 246.6],
    [16, 130, 188, 247],
    [18, 131.4, 189.4, 248.4],
    [20, 132.1, 190.1, 249.1],
])

x, y1, y2, y3 = data.T

bar_width = 0.4  # Reduced bar width for even spacing

fig, ax = plt.subplots(figsize=(12, 6))  # Increase the figure size

# Generate a color gradient
colors = plt.cm.viridis(np.linspace(0, 1, 3))

bar1 = ax.bar(x - bar_width, y1, bar_width, label='20 cells', color=colors[0])
bar2 = ax.bar(x, y2, bar_width, label='30 cells', color=colors[1])
bar3 = ax.bar(x + bar_width, y3, bar_width, label='40 cells', color=colors[2])

ax.set_xlabel("Number of Drivers")
ax.set_ylabel("Total Decryption Time (ms)")
ax.legend(loc='upper left')

ax.set_xticks(x)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each part of the bar plot with its value
for i in range(len(x)):
    ax.text(x[i] - bar_width, y1[i] / 2, f'{y1[i]:.1f}', ha='center', va='center', fontsize=10, color='white')
    ax.text(x[i], (y1[i] + y2[i]) / 2, f'{y2[i]:.1f}', ha='center', va='center', fontsize=10, color='white')
    ax.text(x[i] + bar_width, (y1[i] + y2[i] + y3[i]) / 2, f'{y3[i]:.1f}', ha='center', va='center', fontsize=10, color='white')

ax.set_title("Decryption Time by Number of Drivers", fontsize=14)
ax.set_xlim(0.5, max(x) + 0.5)  # Adjust the x-axis limits for better presentation

# Add a shaded region to emphasize a specific range of values
ax.fill_between(x, 220, 260, color='gray', alpha=0.3, label='Acceptable Range')

# Customize colorbar
sm = plt.cm.ScalarMappable(cmap=plt.cm.viridis, norm=plt.Normalize(vmin=20, vmax=40))
sm.set_array([])
cbar = plt.colorbar(sm)
cbar.set_label("Number of Cells")

plt.savefig("advanced_stacked_bar_plot.png", dpi=300, bbox_inches='tight')  # Save as a high-resolution image

plt.show()
```
