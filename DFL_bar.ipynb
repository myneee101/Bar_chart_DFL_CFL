```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [1, 5421.3311, 5382.6558],
    [2, 5411.078, 5380.1643],
    [3, 5408.652, 5376.3484],
    [4, 5425.75, 5387.43],
    [5, 5416.0202,5380.1944],
])

x = np.arange(len(data))
y1 = data[:, 1].astype(float)
y2 = data[:, 2].astype(float)

bar_width = 0.35

fig, ax = plt.subplots(figsize=(8, 4))

colors = plt.cm.viridis(np.linspace(0, 1, 3))

bar1 = ax.bar(x - bar_width/2, y1, bar_width, label='CFL', color=colors[1], edgecolor='black')
bar2 = ax.bar(x + bar_width/2, y2, bar_width, label='DFL', color=colors[0], edgecolor='black')

ax.set_xlabel("Round")
ax.set_ylabel("Executing time (%)")
ax.set_title("Duration Per Round")
ax.set_xticks(x)
ax.set_xticklabels(data[:, 0], fontsize=7)  # Adjust the fontsize as needed
ax.legend(loc='upper right')
ax.grid(alpha=0.3)

ax.set_xlim(-1, len(data))  # Adjusted x-axis limits for better presentation

plt.savefig("DFL_time_bar.pdf", dpi=300, bbox_inches='tight')

plt.show()
```
